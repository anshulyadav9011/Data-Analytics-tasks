### Remove outliers 

```python
ZScores = stats.zscore(credit_card)
Threshold = 3
outliers = (ZScores > Threshold) | (ZScores < -Threshold)
new_data = credit_card[~outliers.any(axis=1)]
```

```python
plt.figure(figsize=(12,8))
new_data.boxplot()
plt.title("Box plot after removeing almost outliers")
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.show()
```

### visualize the features of dataset

```python
new_data.hist(figsize=(16,8),bins=20)
plt.tight_layout()
plt.show()
```

```python
#time vs amount
plt.figure(figsize=(16,10))

plt.subplot(2, 1, 1)
plt.plot(new_data['Time'],new_data['Amount'],color='green')
plt.title('Time vs Amount')
plt.xlabel("Time")
plt.ylabel("Amount")
```

### Statistics transactions

```python
stats = new_data.agg({
    'Time' : ['mean','median','min','max','std'],
    'Amount' : ['mean','median','min','max','std'],
})
stats
```
